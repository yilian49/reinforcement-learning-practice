```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
baseline = torch.load('baseline.pt')
domian_randomization = torch.load('DR.pt')
```

```python
baseline_data = baseline['training_info']['test_scores']
DR_data = domian_randomization['training_info']['test_scores']
```

```python
gravities = [7, 9, 9.8, 10, 12, 14, 16, 18, 20]
lengths = [0.3, 0.4, 0.5, 0.7, 0.9, 1.1, 1.5, 2.0, 2.5, 3.0]
```

```python
Z_baseline_mean = np.array([[item[0] for item in sublist] for sublist in baseline_data])
Z_DR_mean = np.array([[item[0] for item in sublist] for sublist in DR_data])

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Plotting heatmap for baseline_data
c1 = ax[0].imshow(Z_baseline_mean, cmap='viridis', interpolation='nearest', origin='lower')
ax[0].set_title('Baseline Data')
ax[0].set_xticks(np.arange(len(lengths)))
ax[0].set_yticks(np.arange(len(gravities)))
ax[0].set_xticklabels(lengths)
ax[0].set_yticklabels(gravities)
ax[0].set_xlabel('Length')
ax[0].set_ylabel('Gravity')
fig.colorbar(c1, ax=ax[0])

# Plotting heatmap for DR_data
c2 = ax[1].imshow(Z_DR_mean, cmap='viridis', interpolation='nearest', origin='lower')
ax[1].set_title('DR Data')
ax[1].set_xticks(np.arange(len(lengths)))
ax[1].set_yticks(np.arange(len(gravities)))
ax[1].set_xticklabels(lengths)
ax[1].set_yticklabels(gravities)
ax[1].set_xlabel('Length')
ax[1].set_ylabel('Gravity')
fig.colorbar(c2, ax=ax[1])

plt.tight_layout()
plt.show()
```
